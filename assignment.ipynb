```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#1
x = np.linspace(0,100,num=6)
y = np.linspace(0,200,num=6)

plt.plot(x,y,'b')
plt.xlabel('X')
plt.yticks(np.arange(min(y), max(y)+1, 50.0))
plt.ylabel('Y')
plt.title('title')
plt.show()
```

```python
#2
a = np.linspace(0,1.0,num=6,endpoint='False')
b = a

fig = plt.figure()

axs1 = fig.add_axes([0.1,0.1,0.9,0.8])
axs2 = fig.add_axes([0.3,0.5,0.2,0.2])

plt.xticks(np.arange(min(x), max(x)+1, 0.2))
plt.yticks(np.arange(min(x), max(x)+1, 0.2))
axs1.plot(a,b, lw = 0)
axs2.plot(a,b, lw = 0)

plt.show()
```

```python
#3
x1 = np.linspace(0,100,num=6)
y1 = np.linspace(0,200,num=6)
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.9,0.8])
axes2 = fig.add_axes([0.25,0.5,0.2,0.2])
plt.xticks(np.arange(min(x1), max(x1)+1, 20))
plt.yticks(np.arange(min(y1), max(y1)+1, 50))
axes1.plot(x1,y1,'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')

axes2.plot(x1,y1,'r')
axes2.set_xlabel('x')
axes2.set_ylabel('y')




plt.show()
```
